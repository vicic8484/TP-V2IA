```python
master_dir = "."
images_dir = master_dir + "/images"
```

```python
#imports
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Atelier 1
img1 = cv2.imread(images_dir + "/circuit.bmp", cv2.IMREAD_GRAYSCALE)

print("img1 shape : ", img1.shape)
#cv2.imshow("img1", img1)
#cv2.waitKey(10000)
#cv2.destroyAllWindows()

#img1_croped = img1[0:100, 0:100]
img1_croped_copy = img1.copy()[0:100, 0:100]
#cv2.rectangle(img1_croped,(0,0),(20,20),255)

cv2.rectangle(img1_croped_copy,(0,0),(20,20),255)
cv2.imshow("img1", img1)
#cv2.imshow("img1_croped", img1_croped)
cv2.imshow("img1_croped_copy", img1_croped_copy)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
# Atelier 2
img1 = cv2.imread(images_dir + "/colordot.bmp")
B = img1[:,:,0]
G = img1[:,:,1]
R = img1[:,:,2]
print("img1 shape : ", img1.shape)
cv2.imshow("img1", img1)
cv2.imshow("B", B)
cv2.imshow("G", G)
cv2.imshow("R", R)
cv2.waitKey(10000)
cv2.destroyAllWindows()

histB = cv2.calcHist([B],[0],None,[256],[0,256])
histG = cv2.calcHist([G],[0],None,[256],[0,256])
histR = cv2.calcHist([R],[0],None,[256],[0,256])

plt.plot(histB, color='b')
plt.plot(histR, color='r')
plt.plot(histG, color='g')
plt.show()

img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

```
